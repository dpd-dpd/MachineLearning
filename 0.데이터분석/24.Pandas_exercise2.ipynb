타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category1 열을 만드시오. 
category1 카테고리는 다음과 같이 정의됨
1) 20살이 넘으면 성별을 그대로 사용한다.
2) 20살 미만이면 성별에 관계없이 “child”라고 한다

```python
titanic['category1'] = titanic.apply(lambda x: 'child' if x.age < 20 else x.sex, axis=1)
# titanic['category1'] = titanic.apply(lambda x: x.sex if x.age > 20 else 'child', axis=1)
# 위에껀 나이가 NaN이면 else x.sex니까 성별을 사용
# 밑에껀 else 'child'니까 child로함
titanic.tail()
```

타이타닉호의 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 고객의 평균 나이
값이 되도록 titanic 데이터프레임을 고치시오

```python
titanic.fillna({'age':titanic['age'].mean()}, inplace=True)
# titanic.age = titanic.age.fillna(titanic.age.mean())
```
